```python
# Specify the column indices to remove (e.g., 0 for 'A', 1 for 'B', etc.)
cols_to_remove = [0, 1, 3, 6, 7, 9, 12, 13, 14, 19, 23, 26, 27, 29, 31, 32, 33, 35]

# Drop columns by index
dailyFeatures = dailyFeatures.drop(dailyFeatures.columns[cols_to_remove], axis=1)
```

```python
# Load the dataset with correct header row
streamFeatures = pd.read_csv('athabascaStreamDaily.csv', header=1)

# Keep only rows from index 35806 onwards
streamFeatures = streamFeatures.iloc[35805:]

# Remove the first column
streamFeatures = streamFeatures.drop(streamFeatures.columns[0], axis=1)

# Reset index after slicing
streamFeatures = streamFeatures.reset_index(drop=True)
```

```python
streamFeatures.head(10)
```

```python
streamFeatures.to_csv('modified_athabascaStreamDaily.csv', index=False)
```

```python
dailyFeatures.to_csv('modified_jasperDaily2011.csv', index=False)
```

```python
combined_df = dailyFeatures.join(streamFeatures)
```

```python
combined_df.head(5)
```

```python
# Save the combined DataFrame to a new file
combined_df.to_csv('combined_columns_file.csv', index=False)
```

```python
# Remove columns where all values are NaN
df_cleaned = combined_df.dropna(axis=1, how='all')

# Save the cleaned DataFrame to a new file
df_cleaned.to_csv('cleaned_file.csv', index=False)
```
