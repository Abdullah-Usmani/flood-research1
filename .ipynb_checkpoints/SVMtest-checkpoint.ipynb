```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVR

# Define the parameter distribution
param_dist = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'poly', 'sigmoid']
}

# Set up the Random Search
random_search = RandomizedSearchCV(
    estimator=SVR(), 
    param_distributions=param_dist, 
    n_iter=10,  # Fewer iterations
    cv=3,       # Fewer cross-validation folds
    n_jobs=-1, 
    random_state=42,
    verbose=3   # Increase verbosity to track progress better
)

# Fit the model
random_search.fit(X_train, Y_train)

# Output the best parameters and score
print("Best Parameters:", random_search.best_params_)
print("Best Cross-Validation Score:", random_search.best_score_)

# Make predictions using the best model
best_svr = random_search.best_estimator_
Y_pred = best_svr.predict(X_test)

# Evaluate the performance on test data
r_squared = r2_score(Y_test, Y_pred)
print(f"R-squared: {r_squared}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming grid_search is the result from GridSearchCV or RandomizedSearchCV
results = pd.DataFrame(random_search.cv_results_)

# Pivot the table to get a matrix of mean test scores
pivot_table = results.pivot(index='param_C', columns='param_gamma', values='mean_test_score')

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap='viridis')
plt.title('Grid Search Hyperparameter Tuning')
plt.xlabel('Gamma')
plt.ylabel('C')
plt.show()
```
